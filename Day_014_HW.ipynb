### 請根據不同的 HOUSETYPE_MODE 對 AMT_CREDIT 繪製 Histogram

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件

%matplotlib inline
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')

# 設定 data_path
dir_data = './data/'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.shape
```

## Check row data whose "HOUSETYPE_MODE" is nan

```python
# the index whose "HOUSETYPE_MODE" data is nan
drop_index = app_train[app_train["HOUSETYPE_MODE"].isna()].index
print("drop_index = ", drop_index)

# the index whose "HOUSETYPE_MODE" data is NOT nan
reserve_index = app_train[~(app_train["HOUSETYPE_MODE"].isna())].index
print("reserve_index = ", reserve_index)
```

```python
app_train = app_train.loc[reserve_index, :]
app_train["HOUSETYPE_MODE"].unique()
```

```python
app_train.loc[:, ["HOUSETYPE_MODE", "AMT_CREDIT"]]
```

```python
"""
Your Code Here
"""
unique_house_type = app_train["HOUSETYPE_MODE"].unique()

nrows = len(unique_house_type)
ncols = nrows // 2

plt.figure(figsize=(8, 20))
for i in range(len(unique_house_type)):
    plt.subplot(nrows, ncols, i+1)
    """
    Your Code Here
    """
    app_train.loc[(app_train["HOUSETYPE_MODE"] == unique_house_type[i]), "AMT_CREDIT"].hist()
    
    plt.title(str(unique_house_type[i]))
plt.show()    
```
